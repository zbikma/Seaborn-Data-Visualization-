```python
import pandas as pd
pd.plotting.register_matplotlib_converters
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
from multiprocessing import pool
```

```python
flight_delay=pd.read_csv('flight_delays.csv',index_col='Month')
flight_delay.head()
```

```python
plt.figure(figsize=(20,10))
plt.title("Average Arrival Delay for Sprint Airline Flights, by Month")
plt.xlabel("Month")
plt.ylabel("Arrival Delay (in Minutes)")
sns.barplot(flight_delay.index,y=flight_delay['NK'])
```

```python
plt.figure(figsize=(20,10))
plt.title("Average Arrival Delay for each Airline, by Month")
plt.xlabel("Airline")
sns.heatmap(data=flight_delay,annot=True)
```

## Scatter Chart

```python
# the data set is called insurance.csv We'll work with a (synthetic) dataset of insurance charges, to see if we can understand why some customers pay more than others
import pandas as pd
insurance_data=pd.read_csv('insurance.csv')
insurance_data
```

```python
plt.figure(figsize=(10,10))
plt.title("Insurance charge vs BMI")
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['smoker'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['sex'])
```

```python
plt.figure(figsize=(10,10))
sns.lmplot(x='bmi',y='charges',hue='smoker',data=insurance_data)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x=insurance_data['sex'],y=insurance_data['charges'])
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'])#,hue=insurance_data['sex'])
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'],hue=insurance_data['sex'])
```
